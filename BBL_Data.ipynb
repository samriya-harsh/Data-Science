```python
# Reading in the data
deliveries_df = pd.read_csv("deliveries.csv.zip")
deliveries_df
```

```python
deliveries_df.info()
```

```python
# Sorting the data
deliveries_df = deliveries_df.sort_values(["id", "inning", "over", "ball"], ascending = True)
deliveries_df[121:127]
```

```python
# Batsmen with the most runs
most_runs = deliveries_df.groupby("batsman").sum()
most_runs[["batsman_runs"]].sort_values("batsman_runs", ascending = False)[:10]
```

```python
# Batsmen with the best average having scored more than 1000 runs
most_runs["batting_average"] = round(most_runs["batsman_runs"] / most_runs["is_wicket"], 2)
bat_average = most_runs.sort_values("batting_average", ascending = False)
best_batsmen = bat_average.loc[(bat_average["batting_average"] < np.inf) & (bat_average["batsman_runs"] > 1000)][:10]
best_batsmen[["batsman_runs", "batting_average"]]
```

```python
deliveries_df["extras_type"].unique()
```

```python
# Runs scored off bowler by batsmen only
runs_scored_off_bowler = deliveries_df.groupby("bowler").sum()
runs_scored_off_bowler["batsman_runs"]
```

```python
# Runs conceded bowler via extra. Wides anf no balls only
extra_runs = deliveries_df.loc[(deliveries_df["extras_type"] == "wides") |
                                      (deliveries_df["extras_type"] == "noballs")]
extra_runs_bowler = extra_runs.groupby("bowler").sum()
extra_runs_bowler["extra_runs"]
```

```python
# Runs scored off bowler by batsmen plus extras conceded
total_bowler_runs = runs_scored_off_bowler["batsman_runs"] + extra_runs_bowler["extra_runs"]
total_bowler_runs
```

```python
is_dismissal = deliveries_df.loc[deliveries_df["is_wicket"] == 1]
is_dismissal["dismissal_kind"].unique()
```

```python
dismissal_for_bowler = is_dismissal["dismissal_kind"].str.contains("caught|bowled|lbw|stumped|hit wicket|caught and bowled'")
is_dismissal[dismissal_for_bowler]["dismissal_kind"].unique()
```

```python
# Wickets taken by bowler
wickets = is_dismissal[dismissal_for_bowler].groupby("bowler").sum()
wickets["is_wicket"]
```

```python
# Bowler average
bowler_average = round(total_bowler_runs / wickets["is_wicket"], 2)
bowler_average
```

```python
# Put all data in a DataFrame
bowler_stats_df = pd.DataFrame()
bowler_stats_df["runs"] = total_bowler_runs
bowler_stats_df["wickets"] = wickets["is_wicket"]
bowler_stats_df["average"] = bowler_average
bowler_stats_df
```

```python
# Bowler with the best average having taken more than 50 wickets
bowler_stats_df.loc[bowler_stats_df["wickets"] > 50].sort_values("average", ascending = True)[:10]
```
